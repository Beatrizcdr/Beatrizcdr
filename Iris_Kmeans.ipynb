# Agrupamento do Conjunto de Dados Iris

# Importação das bibliotecas e módulos do Python

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
from sklearn.cluster import KMeans
from sklearn import datasets, cluster, metrics
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder
import seaborn as sns


%matplotlib inline
```

# Importação do Conjunto de Dados

```python
iris = datasets.load_iris()
```

# Agrupamento pelo K-means
Inicializando o K-means e agrupando os dados

```python
#Utilizando diferentes valores de K (agrupamentos):
#Considerando a escolha dos centróides de forma aleatória pelo algoritmo indutor:


resultado = {k: cluster.KMeans(k, random_state=42).fit_predict(iris.data) for k in [2, 3, 4, 6]}
print('\n',"Rótulos gerados pelo K-means:",'\n')
resultado
```

# Representação gráfica
k=2

```python
kmeans = KMeans(n_clusters = 2, random_state=0)
km2 = kmeans.fit_predict(iris.data)
rotulos2 = kmeans.labels_
centroides= kmeans.cluster_centers_

print('\n',"Coordenadas dos centróides:",'\n',centroides,'\n')
print('\n',"Rótulos gerados pelo K-means:",'\n', rotulos)  
```

```python
x= iris.data
y= iris.target
centers = centroides

plt.scatter(x[:, 0], x[:,1], s = 100, c = rotulos2)
plt.scatter(centers[:, 0], centers[:, 1], s = 300, c = 'red',label = 'Centroids')
plt.title('Iris Clusters and Centroids')
plt.xlabel('SepalLength')
plt.ylabel('SepalWidth')
plt.legend()
plt.savefig('kmeans_letraAitem1_CentSepal2.png')
plt.show()
```

# Verificando o k ideal pelo método Elbow

```python
#Lista para agregar o valor do somatório da variância dos dados:
#Considerando a escolha dos centróides de forma aleatória pelo algoritmo indutor:

wcss = []
print ("k",'\t','Somatório dos erros quadráticos das instâncias de cada cluster:')
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i,init = 'random', random_state=0)
    kmeans.fit_predict(iris.data)
    print (i,'\t',kmeans.inertia_)
    wcss.append(kmeans.inertia_)  
plt.plot(range(1, 11), wcss)
plt.title('"K" ideal (Método Elbow)')
plt.xlabel('Número de Agrupamentos')
plt.ylabel('Somatório da Variância dos Dados') #within cluster sum of squares
plt.show()
```
